 Question 3 : Define mean, median, and mode. Explain why these measures of central
tendency are important.


answer: Mean: The arithmetic average of all data points. It is calculated by summing all values and dividing by the total number of observations.

Median: The middle value in a dataset that has been arranged in order (ascending or descending). It splits the data into two equal halves; if the dataset has an even number of values, it is the average of the two middle numbers.


Mode: The value that appears most frequently in a dataset. A dataset can have one mode, multiple modes (bimodal or multimodal), or no mode if all values appear only once.


Question 4: Explain skewness and kurtosis. What does a positive skew imply about the
data?


answer: Skewness

Skewness measures the asymmetry of a probability distribution around its mean. It indicates whether the data is "lopsided" toward one side or the other.

Positive Skew (Right-Skewed): The distribution has a long "tail" extending toward the right (higher values).

Negative Skew (Left-Skewed): The distribution has a long tail extending toward the left (lower values).
Zero Skew: The distribution is perfectly symmetrical, like a normal distribution

Kurtosis

Kurtosis measures the "tailedness" of the distribution—the weight or thickness of the tails relative to the center. It is often used to identify the presence of outliers.

Leptokurtic (High Kurtosis): Heavy tails and a sharp peak, indicating a higher frequency of extreme outliers.

Platykurtic (Low Kurtosis): Light tails and a flatter peak, indicating fewer outliers and a more uniform spread.
Mesokurtic: A normal distribution with a kurtosis of 3 (or an excess kurtosis of 0).


Implications of a Positive Skew
A positive skew (or right-skew) implies specific characteristics about the data:
Tail Direction: The distribution is pulled toward the right, meaning there is a long tail of infrequent, exceptionally high values.

Data Concentration: Most data points are clustered on the left side (lower values), with a few extreme "high" outliers.

Relationship of Measures: The mean is typically greater than the median, which is greater than the mode (Mean > Median > Mode). This happens because the high outliers pull the mean away from the center.

Real-World Example: Household income is a classic example of positive skew; most people earn near the lower or middle range, while a very small number of high earners create a long tail to the right.

Question 5: Implement a Python program to compute the mean, median, and mode of
a given list of numbers.
numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]


```python
import statistics

numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]
mean_val = sum(numbers) / len(numbers)
median_val = statistics.median(numbers)
mode_vals = statistics.multimode(numbers)
print(f"Mean: {mean_val}")
print(f"Median: {median_val}")
print(f"Mode(s): {mode_vals}")
```

Question 6: Compute the covariance and correlation coefficient between the following
two datasets provided as lists in Python:
list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

```python
import numpy as np

list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

x = np.array(list_x)
y = np.array(list_y)

# Returns a 2x2 matrix; [0,1] is the covariance between x and y
cov_matrix = np.cov(x, y)
covariance = cov_matrix[0, 1]
print(covariance)

# Returns the correlation matrix
corr_matrix = np.corrcoef(x, y)
correlation = corr_matrix[0, 1]
print(correlation)
```

Question 7: Write a Python script to draw a boxplot for the following numeric list and
identify its outliers. Explain the result:
data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

```python
import matplotlib.pyplot as plt
import numpy as np

# Dataset
data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

# 1. Calculate Quartiles and IQR
q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)
iqr = q3 - q1

# 2. Define Outlier Bounds (Tukey's Fences)
lower_bound = q1 - (1.5 * iqr)
upper_bound = q3 + (1.5 * iqr)

# 3. Identify Outliers
outliers = [x for x in data if x < lower_bound or x > upper_bound]

# 4. Draw Boxplot
plt.figure(figsize=(8, 5))
plt.boxplot(data, vert=False, flierprops=dict(markerfacecolor='r', marker='o'))
plt.title("Boxplot of Data with Outlier Identification")
plt.xlabel("Value")
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()

print(f"Lower Bound: {lower_bound}, Upper Bound: {upper_bound}")
print(f"Identified Outliers: {outliers}")

```

Question 8: You are working as a data analyst in an e-commerce company. The
marketing team wants to know if there is a relationship between advertising spend and
daily sales.
● Explain how you would use covariance and correlation to explore this
relationship.
● Write Python code to compute the correlation between the two lists:
advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]


```python
import numpy as np

# Data provided
advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

# Compute the correlation matrix
correlation_matrix = np.corrcoef(advertising_spend, daily_sales)

# Extract the correlation coefficient between the two lists
correlation_xy = correlation_matrix[0, 1]

print(f"Correlation Coefficient: {correlation_xy:.4f}")
```

Question 9: Your team has collected customer satisfaction survey data on a scale of
1-10 and wants to understand its distribution before launching a new product.
● Explain which summary statistics and visualizations (e.g. mean, standard
deviation, histogram) you’d use.
● Write Python code to create a histogram using Matplotlib for the survey data:
survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

```python
import matplotlib.pyplot as plt

# Survey data
survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

# Create the histogram
plt.figure(figsize=(8, 5))
# bins=range(1, 12) ensures each score (1-10) has its own bar
plt.hist(survey_scores, bins=range(1, 12), color='skyblue', edgecolor='black', align='left')

# Adding labels and title
plt.title('Distribution of Customer Satisfaction Scores')
plt.xlabel('Satisfaction Score (1-10)')
plt.ylabel('Frequency (Number of Respondents)')
plt.xticks(range(1, 11))  # Ensure all scale points are visible
plt.grid(axis='y', alpha=0.3)

# Display the plot
plt.show()
```
